## Task 1: Basic Python Programming





##### Write a function `fibonacci(n)` that generates the first `n` numbers in the Fibonacci sequence and returns

```python
def fibonacci(n):
    sequence = []
    a, b = 0, 1
    while len(sequence) < n:
        sequence.append(a)
        a, b = b, a + b
    return sequence

n = int(input("Enter the number: "))
print(fibonacci(n))
```

##### Write a function `is_prime(num)` that checks if a given number is a prime number. Return `True` if the number is prime and return 'False' if the number is non-prime

```python
def is_prime(num):
    if num <= 1:
        return False
    if num == 2:
        return True
    if num % 2 == 0:
        return False
    for i in range(3, int(num**0.5) + 1, 2):
        if num % i == 0:
            return False
    return True

num = int(input("Enter a number: "))
print(is_prime(num))
```

## Task 2: Data Manipulation with Pandas

```python
data = {
'Date': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05'],
'Sales': [200, 220, 210, 215, 225],
'Region': ['North', 'South', 'East', 'West', 'North'],
'Manager': ['John', 'Jane', 'Joe', 'Jake', 'John']
}
```

```python
import pandas as pd
df = pd.DataFrame(data)
df
```

##### Load the dataset into a pandas DataFrame and display its first 5 rows.

```python
df.head()
```

##### Add a new column `Profit` which is 10% of the `Sales` column.

```python
df['Profit'] = df['Sales'] * 0.10
df
```

##### Group the data by `Region` and calculate the total sales for each region.

```python
total_sales = df.groupby('Region')['Sales'].sum()
print("\nTotal sales for each region:")
print(total_sales)
```

##### Find the region with the highest average sales.

```python
average_sales_by_region = df.groupby('Region')['Sales'].mean()
highest_avg_sales = average_sales_by_region.idxmax()
print("\nRegion with the highest average sales:")
print(f"{highest_avg_sales} with an average sales of {average_sales_by_region[highest_avg_sales]:.2f}")
```
