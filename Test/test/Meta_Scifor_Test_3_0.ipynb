### Part 1: Pandas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
 'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
 'Age': [24, 27, 22, 32, 29],
 'Score': [85.5, 90.3, 88.0, 92.1, 78.8]
}
```

```python
df1 = pd.DataFrame(data)
df1
```

##### Display three rows

```python
df1.head(3)
```

##### Select and display the 'Name' and 'Score' columns from the DataFrame

```python
df1[['Name', 'Score']]
```

##### Filter the DataFrame to show only the rows where the 'Age' is greater than 25


```python
print("Displaying only those rows where age is greater than 25")
print(df1[df1['Age'] > 25])
```

##### Add a new column 'Passed' to the DataFrame, where the value is True if the 'Score' is greater than 80, otherwise False.


```python
df1['Passed'] = df1['Score'].apply(lambda x: True if x > 80 else False)
print("Displaying the new column 'Passed'")
print(df1)
```

##### Sort the DataFrame by the 'Score' column in descending order


```python
df1.sort_values(by='Score', ascending=False)
```

```python
df1
```

##### Group the DataFrame by the 'Passed' column and calculate the mean 'Score' for each group.


```python
print("Calculating the mean Score for each group")
print(df1.groupby('Passed')['Score'].mean())
```

##### Add a new column 'Attendance' with values [90, np.nan, 85, 87, np.nan]. Write the code to fill missing values with the mean of the column

```python
df1['Attendance'] = [90, np.nan, 85, 87, np.nan]
df1
```

```python
df1['Attendance'].fillna(df1['Attendance'].mean(), inplace=True)
df1
```
