```python
# inserting null values 'NAN' into the dataset
dict_2 = {'A': [1,3,np.nan] , 'B': [np.nan,3,4] , 'C': [3,np.nan,5]}
```

```python
df2 = pd.DataFrame(dict_2)
df2
```

```python
df2['Non NAN Values'] = [2,3,4]
df2
```

```python
# drop nan values from a dataset
# axis = 1 for dropping columns having nan values
# axis = 0 for dropping rows having nan values
print(df2.dropna(axis=1))
```

```python
# no change in original df2 after dropping nan values
df2
```

```python
# filling nan with some values in the dataset
print(df2.fillna(value=9))
```

```python
# Again no change in the original df2
df2
```

```python
# To change the nan values in the original dataset df2, use inplace = True
print(df2.fillna(value=9, inplace=True))
```

```python
data={'Company':['Goog','Goog','MSFT','MSFT','FB','FB'],
      'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
     'Sales':[200,120,340,124,243,350]}
```

```python
df3 = pd.DataFrame(data)
df3
```

```python
# Finding mean company wise
x = df3.groupby('Company')
print(x.mean())
```

```python
# Finding mean company wise, only for numerical columns
x = df3.groupby('Company')['Sales']
print(x.mean())
```

```python
class_a=pd.DataFrame({'Mat':[45,68,35,65],
                     'Eng':[54,87,36,52],
                     'Sci':[66,77,88,70],
                     'Chem':[54,79,58,76]},
                    index=['Ravi','Shyam','Saurav','Abhi'])
class_a
```

```python
class_b=pd.DataFrame({'Mat':[50,87,69,71],
                     'Eng':[66,50,87,69],
                     'Sci':[34,66,69,71],
                     'Chem':[54,87,34,66]},
                    index=['Shaan','Mahi','Viru','Mona'])
class_b
```

```python
class_c=pd.DataFrame({'Mat':[66,50,87,56],
                     'Eng':[56,87,89,71],
                     'Sci':[77,88,70,52],
                     'Chem':[34,56,56,87]},
                    index=['Astha','Rohit','Rahul','Shital'])
class_c
```

```python
# concatenate all three classes row wise using axis = 0
all_class_row = pd.concat([class_a,class_b,class_c],axis=0)
all_class_row
```

```python
# concatenate all three classes column wise using axis = 1
all_class_column = pd.concat([class_a,class_b,class_c],axis=1)
all_class_column
```

```python
# filling na values with 50
all_class_column.fillna(value=50,inplace=True)
all_class_column
```
